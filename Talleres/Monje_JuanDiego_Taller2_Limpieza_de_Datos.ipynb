## 1. Fechas Publicación

Teniendo en cuenta que existe una Fecha de Publicación del Proceso y una fecha de última publicación, realice la limpieza de formato de dichas variables. *Utilice estadísticas descriptivas para responder la pregunta*

```python
df_SECOP['Fecha de Publicacion del Proceso'] = pd.to_datetime(df_SECOP['Fecha de Publicacion del Proceso'], format= '%m/%d/%Y')
df_SECOP['Fecha de Ultima Publicación'] = pd.to_datetime(df_SECOP['Fecha de Ultima Publicación'], format= '%m/%d/%Y')

diff_fecha = df_SECOP['Fecha de Publicacion del Proceso'] - df_SECOP['Fecha de Ultima Publicación']
diff_fecha.describe()
```

## 2. Tipo de Contrato y Subtipo de Contrato
De acuerdo con las variables Contrato y Subcontrato, realice una tabla cruzada y responda ¿Vale la pena dejar ambas variables en la fuente de datos?

```python
pd.crosstab(df_SECOP['Tipo de Contrato'], df_SECOP['Subtipo de Contrato'])
```

**Respuesta:** Al ver la tabla cruzada, se puede ver cómo la variable `Subtipo de Contrato` sólo tiene una categoría: _no especificado_. Siendo así, no se hace necesario mantener esta variable y se decide eliminarla

```python
df_SECOP.drop(columns=['Subtipo de Contrato'],inplace = True)
```

*df_SECOP_copy*  es nuestra nueva base

```python
df_SECOP_copy = df_SECOP[df_SECOP['Unidad de Duracion'] != 'ND']
```

```python
df_SECOP_copy['Unidad de Duracion'].value_counts()
```

```python
# Se tranforma de 'str' a 'int' la variable UNIDAD DE DURACION
df_SECOP_copy.replace({'Unidad de Duracion':{'Dias':1,'Meses':30,'Años':365}},inplace=True)

# Se crea la nueva variable DURACION TOTAL
df_SECOP_copy['Duracion total'] = df_SECOP_copy['Duracion']*df_SECOP_copy['Unidad de Duracion']
```
