## <font color='red'>2. Pollster Ratings - Proyect FiveThirtyEight</font>

**¿Qué?**
- Esta visualización, tambien creada por *five thirty eight*, suele ser algo más especializada, ya que muestra las calificaciones (según una metodología no tan fácil de entender) de todo un universo de encuestadoras que existen en Estados Unidos, ordenadas de peor a mejor, contando con categorías como A+, A, A/B, B+, ..., hasta C/D.

- En general, son datos de tipo cuantitativo, luego categorizados en unas calificaciones estandar de tipo ordinal aunque categóricas.

- **NOTA:** Es dificil compresión, ya que los puntos se acercan mucho uno del otro. Al no tener expresa una calificación cuantitativa, muchas encuestadores se alojan en una misma calificación.



**¿Por qué?**
- El usuario debe explorar cada punto para revizar **diferentes características** como: *encuestas analizadas, años recientes de acción de la encuestadora, afiliación a alguna organización profesional de encuestadoras y, muy importante, el porcentaje de precisión en elecciones pasadas.*

- Se debe saber que no cualquier usuario se sentirá incentivado en revisar, si quiera, al detalle esto. Solo aquellos a quienes les interesa enfocar sus opiniones sobre encuestadoras fiables, medios de comunicación que busca productos estadísticos de calidad o hasta instituciones educativas con ejemplares maneras en materia de investigación estadística.

- Hay una tabla adicional que puede ser chequeada por el usuario, que contiene datos cuantitativos sobre el *margen de predictibilidad*, *las encuestas analizadas totales* y *sesgo promedio de la calificación.* De nuevo, no de interes para todos aquellos que accedan a la visualización.



**¿Cómo?**
- La visualización cuenta con una interacción poco útil, donde sólo si la persona sabe el o los nombres de las encuestadoras a analizar, puede actuar en ella. Una buena interacción podría ser un filtrado por la naturaleza de la firma, sea proveniente de una institución educativa, medios de comunicación o con alguna clasificación con grupos de poder o políticos influyentes.



**MARCAS:** Como ya se mencionó, la desventaja del uso de una marca de *puntos*, al haber un gran número de encuestadoras, hace que se pueda replatear el uso del mismo. Se puede recomendar el filtrado por el número de encuestas anaizadas, dado que entre menos van perdiendo representatividad. 

**CANALES:** Se usan canales de color (paleta de verdes) y forma (círculo/punto). El color de manera acertiva depende de la calificación de cada firma encuestadora.

**Eficacia**
- La clasificación de cada encuestadora se hace de buena manera, al categorizarla por letras y positivos/negativos, sin embargo, sin duda alguna podría ser replanteado. Tal vez también los círculos puedan cambiar, por tamaño o algo por el estilo, quiza espacialmente, para que las calificaciones puedan verse de mejor manera.

- Es pedagógica la visualización, sin embargo, el reto es a mejorarla para que pueda ser expuesta y entendida a todo público, más que ser algo técnico o de caracter investigativo. En suma, hay errores de discriminabilidad, lo que imposibilita captar el grado de precisión de la visualización.
