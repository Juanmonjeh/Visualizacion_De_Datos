## Red 1

Esta red incuye la relación entre weblogs sobre política estadounidense, registrada en 2005 por Adamic y Glance. 
Adamic y N. Glance, "La blogósfera política y las elecciones estadounidenses de 2004"

Fuente: http://www-personal.umich.edu/~mejn/netdata/

```python
pol_books = nx.read_gml('polbooks.gml')
print("Número de nodos:", pol_books.number_of_nodes())
print("Número de nodos:", pol_books.number_of_edges())
```

```python
list(pol_books.nodes)
```

```python
list(pol_books.edges)
```

```python
plt.figure(figsize=(50,20))
plt.subplot(121)
nx.draw(pol_books, with_labels=True)
```
