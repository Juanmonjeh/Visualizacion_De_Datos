**Datos Abortos En Colombia**

```python
import pandas as pd
Abortos = pd.read_excel("C:/Users/juanm/OneDrive/Escritorio/POLITÉ/Respuesta requerimiento IVE.xlsx",sheet_name = "Mun_Prestador")
```

```python
# Eliminando Información basura
Abortos.drop(range(3), inplace = True) 
Abortos.dropna(inplace=True)

# Renombrando Filas y Columnas
Abortos.rename(columns = Abortos.iloc[0], inplace = True)
Abortos.drop(3, inplace = True)
Abortos.rename(columns = {2009.0: 2009, 2010.0: 2010, 2011.0: 2011, 2012.0: 2012, 
                          2013.0: 2013, 2014.0: 2014, 2015.0: 2015, 2016.0: 2016, 
                          2017.0: 2017}, inplace=True)

# Dividiendo las columnas tipo string
Abortos['Coddep'], Abortos['Dep'] = Abortos['Departamento'].str.split('-', 1).str
Abortos['Codmun'], Abortos['Mun'] = Abortos['Municipio'].str.split('-', 1).str
Abortos['Codpres'], Abortos['Pres'] = Abortos['Prestador'].str.split('-', 1).str
Abortos['Coddiag'], Abortos['Diag'] = Abortos['Diagnósticos'].str.split('-', 1).str

# Ordenando columnas de la tabla
Abortos = Abortos[['Coddep','Dep','Codmun','Mun','Codpres','Pres','Coddiag','Diag',
                  2009,2010,2011,2012,2013,2014,2015,2016,2017]]
Abortos = Abortos.astype({2009: int, 2010: int, 2011: int, 2012: int,2013: int, 
                2014: int, 2015: int, 2016: int, 2017.0: int})

Abortos
```

```python
#Solo Abortos Médicos
Abortos_Medicos = Abortos[Abortos['Diag'].str.contains('ABORTO MEDICO')]
Abortos_Medicos.Diag.value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale = 7.5)

fig, axes = plt.subplots(2, 2, figsize=(100, 100), sharey=True)
g1 = sns.barplot(data = Abortos_Medicos, estimator = sum, x = "Coddiag", y = 2014, ci = None, ax = axes[0,0])
g1.set_xlabel('Diagnóstico', fontdict={'fontweight':'bold', 'fontsize':'120'})
g1.set_ylabel('Suma Total', fontdict={'fontweight':'bold', 'fontsize':'120'}) 
g1.set_title('Tipos de Abortos Médicos en 2014', fontdict={'fontweight':'bold', 'fontsize':'140'})
g1.bar_label(g1.containers[0])

g2 = sns.barplot(data = Abortos_Medicos, estimator = sum, x = "Coddiag", y = 2015, ci = None, ax = axes[0,1])
g2.set_xlabel('Diagnóstico', fontdict={'fontweight':'bold', 'fontsize':'120'})
g2.set_ylabel('Suma Total', fontdict={'fontweight':'bold', 'fontsize':'120'}) 
g2.set_title('Tipos de Abortos Médicos en 2015', fontdict={'fontweight':'bold', 'fontsize':'140'})
g2.bar_label(g2.containers[0])

g3 = sns.barplot(data = Abortos_Medicos, estimator = sum, x = "Coddiag", y = 2016, ci = None, ax = axes[1,0])
g3.set_xlabel('Diagnóstico', fontdict={'fontweight':'bold', 'fontsize':'120'})
g3.set_ylabel('Suma Total', fontdict={'fontweight':'bold', 'fontsize':'120'}) 
g3.set_title('Tipos de Abortos Médicos en 2016', fontdict={'fontweight':'bold', 'fontsize':'140'})
g3.bar_label(g3.containers[0])


g4 = sns.barplot(data = Abortos_Medicos, estimator = sum, x = "Coddiag", y = 2017, ci = None, ax = axes[1,1])
g4.set_xlabel('Diagnóstico', fontdict={'fontweight':'bold', 'fontsize':'120'})
g4.set_ylabel('Suma Total', fontdict={'fontweight':'bold', 'fontsize':'120'}) 
g4.set_title('Tipos de Abortos Médicos en 2017', fontdict={'fontweight':'bold', 'fontsize':'140'})
g4.bar_label(g4.containers[0])

fig.savefig('Visualizacion_1_AbortosMedicos4Años.png')
```

## ¿Qué? 

Los datos provienen del **Ministerio de Salud de Colombia**, ordenados en una _tabla_ con un horizonte _temporal_ de 9 AÑOS. Las variables que se logran observar son: **Departamento**, **Municipio**, **Entidad Prestadora**, **Diagnóstico del aborto**, todas estas de tipo _string_ y 9 variables de tipo _numérico_ (2009 a 2017) donde se cuentan la cantidad de abortos. Se espera, para una próxima entrega, actualizar los datos observados. La información es de tipo estática.

## ¿Por qué? 

Con esta visualización se busca identificar al detalle los casos más graves de abortos médicos donde hubo complicaciones como infección (caso incompleto O040, caso completo O045), hemorragia (caso incompleto O041, caso completo O046), embolia (caso incompleto O042, caso completo O047), además de ver los que no tuvieron dificultades.

Objetivo: **Explorar casos** e **Identificar más graves**

## Cómo 

Para ambos casos se es necesario seleccionar los datos únicamente de Abortos Médicos, filtrando la base de datos y agregando la suma total de abortos por cada año.

Select $\rightarrow$ Partition $\rightarrow$ Filter

### Marcas

- **Línea** para el total de abortos médicos por diagnóstico

### Canales

- **Posición horizontal** para ubicar los tipos de abortos médicos
- **Posición vertical** para ubicar la suma total de abortos médicos según el diagóstico
- **Color** para diferenciar cada diagnóstico en los 4 gráficos
- **Posición misma escala vertical** en la suma de totales para guardar interpretabilidad

## Interpretación 

1. Al observar los años seleccionados (2014 a 2017), en el 2016 se registran más de 8mil abortos médicos en todo el país, donde los _abortos médicos completos o no especificado sin complicación_ se ubican cerca de los 6mil, logrando una tasa de más de 16 abortos al día.

2. Para todos los años, se puede decir que los abortos que más predominaron no tuvieron complicaciones, sin embargo los que tuvieron complicaciones no especificadas (O043 incompletos) superan los 100 anuales. Caso atípico para el 2014 que superaban los 200

3. En general, los abortos completos e incompletos con complicaciones especificadas o no, han disminuido a lo largo de los 4 años.

```python
# Totales de Abortos Médicos por Departamento, ordenados
AM_by_dep = Abortos_Medicos.groupby(['Dep']).sum().sort_values([2017],ascending=False).head(10)
AM_by_dep.reset_index(level=0, inplace = True)
```

```python
import matplotlib.pyplot as plt
sns.set()

fig = plt.figure(figsize=(12,12))

plt.subplot2grid((2,2),(0,0))
plt.pie(AM_by_dep[2014], labels = AM_by_dep['Dep'], autopct = '%1.1f%%', pctdistance = 0.85)
plt.title('Abortos Médicos 2014', fontdict = {'fontweight':'bold', 'fontsize':'20'})

plt.subplot2grid((2,2),(0,1))
plt.pie(AM_by_dep[2015], labels = AM_by_dep['Dep'], autopct = '%1.1f%%', pctdistance = 0.85)
plt.title('Abortos Médicos 2015', fontdict = {'fontweight':'bold', 'fontsize':'20'})

plt.subplot2grid((2,2),(1,0))
plt.pie(AM_by_dep[2016], labels = AM_by_dep['Dep'], autopct = '%1.1f%%', pctdistance = 0.85)
plt.title('Abortos Médicos 2016', fontdict = {'fontweight':'bold', 'fontsize':'20'})

plt.subplot2grid((2,2),(1,1))
plt.pie(AM_by_dep[2017], labels = AM_by_dep['Dep'], autopct = '%1.1f%%', pctdistance = 0.85)
plt.title('Abortos Médicos 2017', fontdict = {'fontweight':'bold', 'fontsize':'20'})

plt.show()

fig.savefig('Visualizacion_2_AbortosMedicos4Años.png')
```
